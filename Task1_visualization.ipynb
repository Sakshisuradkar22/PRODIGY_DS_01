```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the World Bank population dataset
data = pd.read_csv('/content/sample_dataset.csv', skiprows=4)  # skip header rows

# Example: Bar chart of population for selected countries
countries = ['India', 'United States', 'China', 'Brazil', 'Germany']
subset = data[data['Country Name'].isin(countries)]
subset = subset.set_index('Country Name')
subset = subset[['2021']]  # choose year
subset.plot(kind='bar', legend=False)
plt.title('Population of Selected Countries (2021)')
plt.ylabel('Population')
plt.show()

# Example: Histogram of all countries' population (2021)
plt.hist(data['2021'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Country Populations (2021)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()
```
